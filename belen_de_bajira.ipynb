### Antioquia aplaudía y celebraba esta decisión. Como no hacerlo, eran varias generaciones antioqueñas que reconocían este territorio como propio. Para otros, era sentir que su identidad cultural le pertenecía legalmente. Sin embargo, el conflicto limítrofe entre Antioquia y Chocó seguía persistiendo. Por lo cual, en el **2014**, el **Instituto Geográfico Agustín Codazzi (IGAC)** inicio la construcción de un informe técnico de deslinde entre ambos departamentos. Cartografías históricas y actuales fueron entregadas en **febrero del 2016** al Congreso de la República de Colombia, quien tendría tan solo un año para pronunciarse, de lo contrario, quedaría como límite provisional el definido por el IGAC en su informe técnico. Estando Belén de Bajirá en el territorio chocoano, de acuerdo a lo establecido en **la ley 13 de 1947** (ley que otorgo el nacimiento al departamento del Chocó). 


### Las Comisiones Conjuntas de Ordenamiento Territorial de Senado y Cámara de Representantes, concluyeron que los estudios cartográficos del IGAC eran confusos y los devolvieron. A pesar de esto, en diciembre del mismo año, estas mismas entidades determinaron que no estaba en su deber ser, definir los límites entre ambos departamentos, para este este caso en particular, era potestad del IGAC dar solución a la disputa limítrofe, el cual culmino **el 10 de junio de 2017** con la publicación de mapas actualizados de ambos departamentos, en el que, **finalmente…Belén de Bajirá pertenecía al Chocó.**

### **Ahora, convertimos el CSV en un DataFrame. Te estas preguntado ¿por qué? Bueno, es necesario para poder manipular la base de datos**

```
dataframe_antioquia = pd.read_csv('dpto_antioquia.csv')
print(dataframe_antioquia)
```

### *Hemos creado datos tabulares que te permite visualizar los elementos que la integran, 125 Filas y 2 Columnas*  

### **Antioquia tiene 125 Municipios. ¡Bastantes! ¿Quieres saber cuáles son los de mayor y menor tamaño?**

> Comencemos creando 2 nuevas columnas. Ya veras lo fácil y rápido que es



```
categoria_municipios= []

for i,row in dataframe_antioquia.iterrows():
  if row['Extensión  km²'] < 1000:
    categoria_municipios.append('Municipio Pequeño')
  elif row['Extensión  km²'] >= 1000 and row['Extensión  km²'] <=2000:
    categoria_municipios.append('Municipio Mediano')
  else:
    categoria_municipios.append('Municipio Grande')


color_tamaño_municipios= []

for i,row in dataframe_antioquia.iterrows():
  if row['Extensión  km²'] < 1000:
    color_tamaño_municipios.append('lightgreen')
  elif row['Extensión  km²'] >= 1000 and row['Extensión  km²'] <=2000:
    color_tamaño_municipios.append('orange')
  else:
    color_tamaño_municipios.append('blue')

dataframe_antioquia['categoria_municipios'] = categoria_municipios
dataframe_antioquia['color_tamaño_municipios'] = color_tamaño_municipios

print(dataframe_antioquia)
```



### **> Construyamos impactantes gráficas con Matplotlib**

###  *Debido al número de Minicipios que tiene Antioquia, fue necesario hacer 5 gráficas, cada una con la información de 25 municipios que estan ordenados alfabéticamente.*
